N-gramas (n-gramas)

```python
frase = "ele sabia que eu sabia trabahar com IA e sabia que eu estava estudando tecnologia sabia disso"
```

```python
# O split transforma a frase ou qualquer texto em lista

palavras = frase.split()
palavras
```

```python
import pandas as pd
tabela = pd.DataFrame({"palavras":palavras})
tabela
```

```python
tabela["proxima_palavra"] = tabela["palavras"].shift(-1)
tabela
```

```python
tabela = tabela.dropna()
tabela
```

```python
filtro = tabela["palavras"] == "eu"
tabela[filtro]
```

```python
tabela[filtro]["palavras"].value_counts()
```

```python
tabela[filtro]["proxima_palavra"].value_counts(normalize=True)
```

```python
tabela[filtro]["proxima_palavra"].value_counts().idxmax()
```
