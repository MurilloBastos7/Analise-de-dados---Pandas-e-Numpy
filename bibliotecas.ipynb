```python
# Importação de bibliotecas
import numpy as np # np sera o apelido
```

```python
temperaturas = np.array([30, 40, 50, 60, 70, 80, 90, 100])
print(temperaturas)
print(type(temperaturas))
```

```python
# Media aritmetica
media = np.mean(temperaturas)
print(f"{media:.2f}")
```

```python
# Maior valor
maior_valor = np.max(temperaturas)
print(maior_valor)
```

```python
# Menor valor
menor_valor = np.min(temperaturas)
print(menor_valor)
```

Desafio final
- criar um array de habitantes de 5 cidades do estado de SP
- maior população 
- menor população
- mediana

```python
import numpy as sp 
```

```python
cidades = sp.array([11451999, 1291771, 1139047,810729, 748919])

print(cidades)

maior_pop = sp.max(cidades)
print(f"A maior é: {maior_pop}")

menor_pop = sp.min(cidades)
print(f"A menor é: {menor_pop}")

mediana_pop = sp.median(cidades)
print(f"A mediana é: {mediana_pop}")

media_pop = sp.mean(cidades)
print(f"A media é: {media_pop}")
```
