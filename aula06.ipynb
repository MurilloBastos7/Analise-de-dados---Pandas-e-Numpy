## Seleção de colunas: Foco em variáveis numéricas para análise de correlação

- **corr():** Calcula a matriz de correlação de Pearson entre todas as colunas selecionadas
- **heatmap():** Cria um mapa de calor para visualizar as correlações de forma intuitiva

Parâmetros:

- **annot=True:** Exibe os valores de correlação nos quadrados
- **cmap="coolwarm":** Esquema de cores com azul para correlações negativas e vermelho para positivas
- **fmt=".2f":** Formata os números com 2 casas decimais

````python
sns.heatmap(data=df[colunas].corr(), annot=True, cmap="coolwarm",fmt=".2f");
````

## 🧮 **[Comparação de médias e exportação](ca://s?q=Comparacao_de_medias_e_exportacao_de_dados)**

## Cálculo de médias e exportação do dataset tratado



- **Filtro 1:** Seleciona estudantes que estudam 8 ou mais horas por dia (filtro_horas_altas)

- **Filtro 2:** Seleciona estudantes que estudam menos de 2 horas por dia (filtro_horas_baixas)

- **Cálculo de médias:** Calcula a média das notas para cada grupo

### Resultado:

Estudantes com ≥ 8 horas: Média = **100.00**

Estudantes com < 2 horas: Média = **45.56**

Análise:
Diferença significativa: Mais de 54 pontos de diferença entre os grupos

- **Evidência clara** O tempo de estudo é um forte preditor do desempenho acadêmico

- **Limitação:** A média de 100.00 no grupo de horas altas sugere que pode haver um teto no dataset ou a amostra é pequena

````python
filtro_horas_altas = df["study_hours_per_day"] >= 8
filtro_horas_baixas = df["study_hours_per_day"] < 2

media_alta = df[filtro_horas_altas]["exam_score"].mean()
media_baixa = df[filtro_horas_baixas]["exam_score"].mean()

print(f"A média das pessoas que estudaram 5 ou mais horas foi {media_alta:.2f}, ja a média das pessoas que estudaram menos de duas horas foi de {media_baixa:.2f}")
````
