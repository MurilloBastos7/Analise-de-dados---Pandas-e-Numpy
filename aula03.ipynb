````python
import pandas as pd
import numpy as np
````

````python
quantidades_produtos = [3,5,6,10]
precos_produtos = [10,5,7,100]
nomes_produtos = ["Pastel", "Caçulinha", "Refrigerante", "Combos"]

array_produtos = np.array(quantidades_produtos)
array_precos = np.array(precos_produtos)

faturamento = array_produtos * array_precos
for nome,preco in zip(nomes_produtos,faturamento):
    print(f"{nome} : {preco}")
````



### Exercicios de fixação 
- Criar uma lista que ira possuir 10 nomes de produtos
- Criar uma lista que ira possuir o preço de custo (atacado)
- Criar uma lista que ira possuir o preço de venda (varejo)
- Ao final, exibir o lucro obtido em cada produto

````python
import numpy as n
import pandas as p

nome_produto = ["maça", "banana", "pera", "uva", "whey", "creatina", "pre-treino", "trembolona", "ovo", "manga"]
preco_atc = [1.50, 1, 2, 0.50, 60, 80, 40, 300, 2, 0.75]
preco_vj = [0.50, 1, 2, 0.50, 50, 70, 30, 250, 1, 0.50]

array_atc = n.array(preco_atc)
array_vj = n.array(preco_vj)

lucro = array_atc - array_vj
for nomes,precos in zip(nome_produto,lucro):
    print(f"{nomes}: {precos}")

print(n.sum(lucro))
````
