Condicionais

```python
carga_horaria = 40
if carga_horaria > 40:
    print("usar uniforme")
else:
    print("não precisa usar uniforme")
```

```python
# solicitar a informação 
carga_horaria = int(input("digite a carga horaria "))
if carga_horaria > 40:
    print("usar uniforme")
else:
    print("usar cracha")
```

```python
avaliacao = int(input("Digite sua avaliação: de 0 - 10 estrelas"))

if avaliacao > 8:
    print("Nota exelente")
elif avaliacao >= 6:
    print("Avaliação regular")
elif avaliacao >= 4:
    print("Avaliação ruim")
else:
    print("Avaliação pessima")
```

```python
nome_filme = input("digite o nome do filme")
print(nome_filme.title())

avaliacao_filme = int((input("Digite a avaliação do filme de 0 - 5 ")))
if avaliacao_filme <= 0 or avaliacao_filme > 5:
    print("Nota invalida!")
print(f"A avaliação é {avaliacao_filme}")
if avaliacao_filme > 5:
    print("Vencedor do Oscar")
elif avaliacao_filme > 3:
    print("Avaliação boa")
else:
    print("Não recomendado")
```
