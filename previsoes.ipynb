```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv(r".\data\base_estudantes_tratado.cvs")
```

```python
df.info()
```

```python
sns.lmplot(data=df,x="study_hours_per_day",y="exam_score");
```

```python
X = df[["study_hours_per_day","sleep_hours","netflix_hours","social_media_hours"]]
y = df["exam_score"]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
modelo = LinearRegression()
modelo.fit(X,y)
```

```python
df.info()
```

```python
novo_aluno = pd.DataFrame({
    "study_hours_per_day":[0],
    "sleep_hours":[1],                
    "netflix_hours":[2],
    "social_media_hours":[3]
    })

modelo.predict(novo_aluno)
```

```python
modelo.coef_
```

```python
modelo.intercept_
```

```python
modelo.score(X,y)
```
